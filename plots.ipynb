```python
from earthquakes import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = get_data()
```

```python
years = []
frequency = []

for earthquake in data["features"]:
    if get_year(earthquake) in years:
        index = years.index(get_year(earthquake))
        frequency[index] += 1
    else:
        years.append(get_year(earthquake))
        frequency.append(1)
```

```python
fig, ax = plt.subplots()
ax.bar(years, frequency)
ax.set_xlabel("Year")
ax.set_ylabel("Number of Earthquakes")
ax.set_title("Number of Earthquakes per Year")
plt.xticks(rotation=45)
plt.yticks(range(0, max(frequency)+1, 2))
plt.show()
```

```python
tot_mag = list([0 for year in years])

for earthquake in data["features"]:
    index = years.index(get_year(earthquake))
    tot_mag[index] += get_magnitude(earthquake)
```

```python
av_mag = np.array(tot_mag)/np.array(frequency)
av_mag = round(av_mag,2)
```

```python
fig, ax = plt.subplots()
ax.bar(years, av_mag)
ax.set_xlabel("Year")
ax.set_ylabel("Average Magnitude")
ax.set_title("Average Magnitude of Earthquakes per Year")
plt.xticks(rotation=45)
plt.show()
```
